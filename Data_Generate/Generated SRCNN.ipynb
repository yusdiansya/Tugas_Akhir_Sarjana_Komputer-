```python
import torch
import numpy as np
from utils import *
from PIL import Image, ImageDraw, ImageFont
from pathlib import Path
import torchvision.transforms as transforms

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
os.makedirs('Generated SRCNN No Label Upscale X4/Generated SRCNN Vid Unseen No Label Epoch 1000 Bt 8 Upscale x4', exist_ok=True)
data_folder = 'Data Mentah (Low Resolution)/Json Vid Unseen/'
out_srresnet = 'Generated SRCNN No Label Upscale X4/Generated SRCNN Vid Unseen No Label Epoch 1000 Bt 8 Upscale x4/'
srgan_checkpoint = "Model SRCNN No Label Upscale X4/Model SRCNN Epoch 1000 Bt 8 Upscale x4_SRCNN.pth.tar"
```

```python
# Load models
# srresnet = torch.load(srresnet_checkpoint)['model'].to(device)
# srresnet.eval()
zoom_factor = 4
srcnn_generator = torch.load(srgan_checkpoint)['model'].to(device)
srcnn_generator.eval()
```

```python
cuda = True
def visualize_sr(img, halve=False):
    img = Image.open(img).convert('YCbCr')
    img = img.resize((int(img.size[0]*zoom_factor), int(img.size[1]*zoom_factor)), Image.BICUBIC)  # first, we upscale the image via bicubic interpolation
    y, cb, cr = img.split()

    img_to_tensor = transforms.ToTensor()
    input = img_to_tensor(y).view(1, -1, y.size[1], y.size[0])  # we only work with the "Y" channel

    #device = torch.device("cuda:0" if (torch.cuda.is_available() and cuda) else "cpu")
    print(device)
    
    input = input.to(device)    
    out = srcnn_generator(input)
     
    out_img_y = out[0].detach().cpu().numpy()
    out_img_y *= 255.0
    out_img_y = out_img_y.clip(0, 255)
    out_img_y = Image.fromarray(np.uint8(out_img_y[0]), mode='L')

    out_img = Image.merge('YCbCr', [out_img_y, cb, cr]).convert('RGB')
    
    #sr_img_srcnn = srcnn_generator(convert_image(img, source='pil', target='imagenet-norm').unsqueeze(0).to(device))
    #sr_img_srcnn = sr_img_srcnn.squeeze(0).cpu().detach()
    #sr_img_srcnn = convert_image(sr_img_srcnn, source='[-1, 1]', target='pil')
    
    
    
#     model = torch.load(model).to(device)
#     input = input.to(device)

#     out = model(input)
#     out = out.cpu()
#     out_img_y = out[0].detach().numpy()
#     out_img_y *= 255.0
#     out_img_y = out_img_y.clip(0, 255)
#     out_img_y = Image.fromarray(np.uint8(out_img_y[0]), mode='L')

#     out_img = Image.merge('YCbCr', [out_img_y, cb, cr]).convert('RGB')  # we merge the output of our network with the upscaled Cb and Cr from before
#                                                                         # before converting the result in RGB
#     out_img.save(f"zoomed_{image}")
    
    return out_img

```

```python
with open(os.path.join(data_folder, 'train_images.json'), 'r') as j:
    src_images = json.load(j)


for src_img in enumerate(src_images):
    visualize_sr(src_img[1]).save(out_srresnet +Path(src_img[1]).stem+'.jpg')
    print('Exporting file No. {0}. from [{1}] to [{2}]'.format(src_img[0]+1, src_img[1], 'a'+Path(src_img[1]).stem+'.jpg'))
```
